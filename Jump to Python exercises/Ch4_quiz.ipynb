### Q2 입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성해 보자.
(단 입력으로 들어오는 수의 개수는 정해져 있지 않다.)

````python
def cal_avg(*args):  # 입력값이 개수를 모를 때 *매개변수
    result = 0
    for num in args:
        result = result + num
    return result / len(args)
````

````python
result = cal_avg(1,2,3,4,5,6)
print(result)
````
